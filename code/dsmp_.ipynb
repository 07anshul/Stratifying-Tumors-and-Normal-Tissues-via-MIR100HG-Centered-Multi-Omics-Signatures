### Median based high and low MIR100HG

```python
PAAD_median_MIR100HG = PAAD_MIR100HG_values.iloc[0].median()

PAAD_median_MIR100HG
```

```python
PAAD_high_low_MIR100HG = pd.DataFrame(columns=['Patient', 'Gene Expression Value', 'MIR100HG Level'])

for i, j in zip(PAAD_MIR100HG_values, PAAD_MIR100HG_values.iloc[0]):
    row_index = len(PAAD_high_low_MIR100HG)
    PAAD_high_low_MIR100HG.at[row_index, 'Patient'] = i
    PAAD_high_low_MIR100HG.loc[row_index, 'Gene Expression Value'] = j
    PAAD_high_low_MIR100HG.loc[row_index, 'MIR100HG Level'] = 'High' if j > PAAD_median_MIR100HG else 'Low'

PAAD_high_low_MIR100HG
```

```python
PAAD_high_MIR100HG = PAAD_high_low_MIR100HG[PAAD_high_low_MIR100HG['MIR100HG Level'] == 'High']
PAAD_low_MIR100HG = PAAD_high_low_MIR100HG[PAAD_high_low_MIR100HG['MIR100HG Level'] == 'Low']

PAAD_high_MIR100HG
```

```python
PAAD_high_MIR100HG_series = PAAD_high_MIR100HG["Patient"]
PAAD_low_MIR100HG_series = PAAD_low_MIR100HG["Patient"]

PAAD_high_MIR100HG_series
```

```python
PAAD_high_MIR100HG.shape, PAAD_low_MIR100HG.shape
```

```python
from scipy.stats import ttest_ind
from statsmodels.stats.multitest import multipletests

def deg_extraction(gene_expression_levels, high_MIR100HG_series, low_MIR100HG_series):
    results = []
    
    gene_expression_levels_ttest = gene_expression_levels.dropna(subset=["HGNC_symbol"])
    hgnc_symbol_list = gene_expression_levels_ttest["HGNC_symbol"]
    gene_expression_levels_ttest = gene_expression_levels_ttest.drop(columns=["HGNC_symbol"])
    for ensembl_id, hgnc_symbol in zip(gene_expression_levels_ttest.index, hgnc_symbol_list):
        if ensembl_id == "ENSG00000255248":
            continue
    
        high_expr = gene_expression_levels_ttest.loc[ensembl_id, high_MIR100HG_series]
        low_expr = gene_expression_levels_ttest.loc[ensembl_id, low_MIR100HG_series]
    
        high_expr = pd.to_numeric(high_expr, errors='coerce')
        low_expr = pd.to_numeric(low_expr, errors='coerce')
    
        high_expr = high_expr.dropna()
        low_expr = low_expr.dropna()

        
        variance_threshold = 0.1
        if len(high_expr) > 1 and len(low_expr) > 1:
            if high_expr.std() > variance_threshold and low_expr.std() > variance_threshold:
                stat, p_value = ttest_ind(high_expr, low_expr, equal_var=False)
                results.append([ensembl_id, hgnc_symbol, stat, p_value])
    
    results_df = pd.DataFrame(results, columns=["Ensembl_id", "HGNC_symbol", "T-statistic", "P-value"])

    return results_df
```

```python
PAAD_results_df = deg_extraction(PAAD_gene_expression_levels, PAAD_high_MIR100HG_series, PAAD_low_MIR100HG_series)
```

```python
PAAD_results_df["Adjusted P-value"] = multipletests(PAAD_results_df["P-value"], method="fdr_bh")[1]

PAAD_results_df["Adjusted P-value"] = pd.to_numeric(PAAD_results_df["Adjusted P-value"], errors="coerce")

PAAD_results_df = PAAD_results_df.sort_values(by="Adjusted P-value", ascending=True)

PAAD_significant_degs = PAAD_results_df[(PAAD_results_df["Adjusted P-value"] < 0.01)]

PAAD_significant_degs = PAAD_significant_degs.sort_values(by="T-statistic", ascending=False)

PAAD_significant_degs.head(20), PAAD_significant_degs.shape
```

```python
PAAD_upregulated_degs = PAAD_significant_degs.sort_values(by='T-statistic', ascending=False).head(100)
PAAD_upregulated_degs.head()
```

```python
PAAD_downregulated_degs = PAAD_significant_degs.sort_values(by='T-statistic').head(100)
PAAD_downregulated_degs.head()
```

```python
PAAD_top_degs = pd.concat([PAAD_upregulated_degs, PAAD_downregulated_degs])

PAAD_top_degs.head(20), PAAD_top_degs.shape
```

```python
def read_degs(file_path, priority='Top'):
    deg_df = pd.read_csv(file_path)
    significant_degs = deg_df[(deg_df["Adjusted P-value"] < 0.01)]
    significant_degs = significant_degs.sort_values(by="T-statistic", ascending=False)
    significant_genes = significant_degs["HGNC_symbol"]
    if priority == "Top":
        upregulated_degs = significant_degs.sort_values(by='T-statistic', ascending=False).head(100)
        downregulated_degs = significant_degs.sort_values(by='T-statistic').head(100)
        top_degs = pd.concat([upregulated_degs, downregulated_degs])
        top_genes = top_degs["HGNC_symbol"]
        return top_degs

    return significant_genes
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.ensemble import IsolationForest
import matplotlib.pyplot as plt

def pca_comparision(combined_samples, keyword):
    scaler = StandardScaler()
    combined_samples_scaled = scaler.fit_transform(combined_samples)

    n_clusters = 2

    #for cancer features
    iso_forest_cancer = IsolationForest(contamination=0.02, random_state=42)
    outlier_labels = iso_forest_cancer.fit_predict(combined_samples_scaled)
    inlier_mask = outlier_labels == 1
    combined_samples_scaled = combined_samples_scaled[inlier_mask]

    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    labels = kmeans.fit_predict(combined_samples_scaled)

    pca = PCA(n_components=2)
    combined_pca = pca.fit_transform(combined_samples_scaled)
    
    print(f"{keyword} Cancer vs Normal PCA% : {pca.explained_variance_ratio_}")
    
    plt.figure(figsize=(8, 6))
    plt.scatter(combined_pca[:, 0], combined_pca[:, 1], c=labels, cmap="viridis")
    plt.title(f"PCA of {keyword} Cancer + Normal Samples (DEG Features)")
    plt.xlabel("PC1")
    plt.ylabel("PC2")
    plt.colorbar(label="Cluster")
    plt.savefig(f"../project/dsmp-2024-groupm22/plots/{keyword}_VS_normal_PCA_plot.png", dpi=300, bbox_inches='tight')
    plt.show()

```

```python

pca_comparision(combined_pancreas_samples, "PAAD")
```
